```python
import pandas as pd

csv = '/Users/banazsinjary/Desktop/871/PoseProject/mpii_human_pose.csv'
df = pd.read_csv(csv)

print(df.describe)
```

```python
sports_df = df[df['Category']=='sports']

print(sports_df.describe())
```

```python
category_df = df.groupby('Category').sum()

print(category_df)
```

```python
df_cleaned = df.drop(columns=[col for col in df.columns if col.startswith('r') or col.startswith('l') or col.startswith('pelvis') or col.startswith('thorax_X')or col.startswith('thorax_Y') or col.startswith('upper')or col.startswith('head')])

print(df_cleaned.head())
```

```python
category_activities = df.groupby('Category')['Activity'].unique()

for category, activities in category_activities.items():
    print(f"Category: {category}")
    print(f"Number of unique activities: {len(activities)}")
    print("Activities:")
    for activity in activities:
        print(f" - {activity}")
    print()
```

```python
if 'Category' in df_cleaned.columns:
    category_counts = df_cleaned.groupby('Category').size()
    print("Category Counts:")
    print(category_counts)
```
